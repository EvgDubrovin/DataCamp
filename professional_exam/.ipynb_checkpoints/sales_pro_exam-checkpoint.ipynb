## 2.4. Was there any difference in revenue over time for each of the methods?

```python
summarised_data = df.groupby(['week', 'sales_method']).revenue.sum().reset_index()

sns.lineplot(data=summarised_data, x='week', y='revenue', hue='sales_method', hue_order=method_order, linewidth=5)

plt.title('Total Revenue Over Time per Approach')
plt.xlabel('Week')
plt.ylabel('Revenue')

for method in method_order:
    plt.text(1, summarised_data[(summarised_data.week==1) & (summarised_data.sales_method==method)].revenue, 
             str(int(summarised_data[(summarised_data.week==1) & (summarised_data.sales_method==method)].revenue.values[0])), 
             va='center', ha='right', fontsize=11)
    plt.text(6, summarised_data[(summarised_data.week==6) & (summarised_data.sales_method==method)].revenue, 
             str(int(summarised_data[(summarised_data.week==6) & (summarised_data.sales_method==method)].revenue.values[0])), 
             va='bottom', ha='left', fontsize=11)

plt.xticks(fontsize=14)
plt.show()
```

```python
# Revenue forecast
from sklearn.linear_model import LinearRegression

# Define the existing data
data = {
    'week': [1, 1, 1, 2, 2, 2, 3, 3, 3, 4, 4, 4, 5, 5, 5, 6, 6, 6],
    'sales_method': ['Call', 'Email', 'Email + Call', 'Call', 'Email', 'Email + Call', 'Call', 'Email', 'Email + Call',
                     'Call', 'Email', 'Email + Call', 'Call', 'Email', 'Email + Call', 'Call', 'Email', 'Email + Call'],
    'revenue': [26796.8190353, 246381.07464608, 19080.75871867, 35098.46046818, 148703.72539231, 30801.11456681,
                37662.539658, 106807.31202279, 54039.20889353, 51662.41669852, 116957.61651315, 89660.33327997,
                55484.65344796, 80846.63972824, 147489.58274151, 29642.26684269, 25789.97123766, 132757.41503639]
}

# Create a DataFrame from the data
temp = pd.DataFrame(data)

# Pivot the data to have week as rows, sales_method as columns, and revenue as values
pivoted_data = temp.pivot(index='week', columns='sales_method', values='revenue').reset_index()

# Fill any missing values with 0
pivoted_data.fillna(0, inplace=True)

# Split the data into features (X) and target (y)
X = pivoted_data['week'].values.reshape(-1, 1)
y_call = pivoted_data['Call'].values
y_email = pivoted_data['Email'].values
y_email_call = pivoted_data['Email + Call'].values

# Create separate models for each sales method
model_call = LinearRegression()
model_email = LinearRegression()
model_email_call = LinearRegression()

# Fit the models
model_call.fit(X, y_call)
model_email.fit(X, y_email)
model_email_call.fit(X, y_email_call)

# Create the future data for prediction
future_weeks = [7, 8, 9, 10, 11, 12]
future_data = pd.DataFrame({'week': future_weeks})

# Perform predictions for each sales method
future_data['Call'] = model_call.predict(future_data[['week']])
future_data['Email'] = model_email.predict(future_data[['week']])
future_data['Email + Call'] = model_email_call.predict(future_data[['week']])
```

```python
# Use melt to combine the 'Call', 'Email', 'Email + Call' columns into a new 'revenue' column
melted_data = pd.melt(future_data, id_vars='week', value_vars=['Call', 'Email', 'Email + Call'],
                      var_name='sales_method', value_name='revenue')

df_forecast = pd.concat([summarised_data, melted_data]).reset_index()
sns.lineplot(data=df_forecast, x='week', y='revenue', hue='sales_method', hue_order=method_order, linewidth=4)

plt.title('Approximate Revenue Forecast - Next 6 Weeks')
plt.xlabel('Week')
plt.ylabel('Revenue')

for method in method_order:
    plt.text(1, df_forecast[(df_forecast.week==1) & (df_forecast.sales_method==method)].revenue, 
             str(int(df_forecast[(df_forecast.week==1) & (df_forecast.sales_method==method)].revenue.values[0])), 
             va='center', ha='right', fontsize=11)
    plt.text(12, df_forecast[(df_forecast.week==12) & (df_forecast.sales_method==method)].revenue, 
             str(int(df_forecast[(df_forecast.week==12) & (df_forecast.sales_method==method)].revenue.values[0])), 
             va='bottom', ha='left', fontsize=11)

plt.xticks(fontsize=14)

# Add a rectangle representing the forecast
plt.axvspan(6, 12, color='gray', alpha=0.2)
plt.text(6.6, plt.ylim()[1]*0.85, 'F      O      R      E      C      A      S      T', color='red')

plt.show()
```

According to the (very) approximate forecast for the next 6 weeks, Email revenue should be negative (which is impossible) and Email + Call revenue should continue to grow up to 310 674. The Call revenue continues to slightly increase.

```python
# Change of performance per each method, in %
round(100*(summarised_data[(summarised_data.week==6) & (summarised_data.sales_method=='Call')].revenue.values[0] - 
           summarised_data[(summarised_data.week==1) & (summarised_data.sales_method=='Call')].revenue.values[0]) / 
           summarised_data[(summarised_data.week==1) & (summarised_data.sales_method=='Call')].revenue.values[0], 2)
```

While the **Email** strategy was the **most successful** one on the 1st week (246 381) and **until the 4th week**, the total revenue decreased by 10x times (89.5%) by the 6th week (25 789). 

Otherwise, the **Email + Call** strategy showed the **best performance** at a distance, becoming the leader by the end of a given period (1st week - 19 080; 6th week - 132 757; 596% growth).

The **Call** method is still an **outsider** (1st week - 26 796; 6th week - 29 642; 10.6% growth).

Since our main goal is to  determine the best selling approach, I would recommend to use these metrics:
1. **Return on Investment (ROI)**:  
***ROI = (Revenue - Cost) / Cost***, where  
* *Revenue* is the total revenue generated by the sales method over a specific period (e.g. next 6 weeks).
* *Cost* is the total cost incurred in implementing and executing the sales method over the same period
* ***Cost = (Average Salary * Average Time Spent) + Average Method Expenses***, where 
    * *Average Salary* is the average salary of a marketing specialist (`$31.36` according to the [U.S. Bureau of Labor Statistics (BLS)](https://www.bls.gov)).
    * *Average Time Spent* is the average time spent on each customer for the specific sales method (should be provided by the Sales team). 
    * *Average Method Expenses* is the ad creation cost, the method development cost, cost of planning, and other related tasks (should be provided by the Sales or Management team).
    
2. **Method Efficiency**:
***Efficiency = Total Revenue * Conversion Rate***, where
* *Conversion Rate* is the percentage of leads that convert into actual customers: ***Conversion Rate = Customers / Total Leads*** for each method.
    * *Total Leads* should be provided by the Sales team.
